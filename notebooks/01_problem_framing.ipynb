## Phase 1 — Problem Framing (Spoken Grammar Quality)

This notebook documents the **assessment-science framing** of the grammar scoring engine for the SHL Intern Hiring Assessment.

### Construct definition

- **Primary construct**: spoken grammatical accuracy and appropriateness in work-relevant English.
- Focus on morphosyntax (tense/aspect, subject–verb agreement, pronouns, prepositions, word order) and sentence/clause well-formedness in spontaneous responses.

### What is not being measured

- Overall job performance or potential.
- General cognitive ability (reasoning, problem-solving).
- Personality, motivation, or socio-emotional skills.
- Content quality, factual accuracy, or domain knowledge.
- Accent quality or pronunciation as primary constructs.
- Adherence to a specific prestige dialect beyond what is needed for mutual intelligibility in an international business context.

### Intended assessment use

- One component in a multi-method assessment battery for intern/early-career roles requiring English communication.
- Supports screening/triage and, where job analysis supports it, minimum proficiency thresholds for spoken grammar.
- Should never be the sole basis for high-stakes decisions.

### Automation risks and ethics

- **ASR bias** across accents/dialects and recording conditions.
- **Construct underrepresentation** (focusing on a subset of grammar features) and **contamination** (prosody, audio quality, content).
- **Fairness** and potential subgroup impact when ASR or models behave differently across candidate groups.
- **Transparency and contestability** of scores (ability to explain and challenge outcomes).
- **Privacy and data governance** for voice data.

This notebook intentionally contains **no code**; it serves as a conceptual anchor for all subsequent technical work.
